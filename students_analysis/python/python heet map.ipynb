```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
columns_efact = [
    "age",
    "study_hours_per_day",
    "social_media_hours",
    "netflix_hours",
    "attendance_percentage",
    "sleep_hours",
    "exercise_frequency",
    "mental_health_rating",
    "exam_score"
]
```

```python
df=pd.read_csv("/content/student_habits_performance.csv")
columns_efact=df[columns_efact]
```

```python
columns_efact.head()
```

```python
corr = columns_efact.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm',fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
